```python
# Chia AGE thành 3 nhóm
boston_df['AGE_group'] = pd.cut(boston_df['AGE'], bins=3, labels=['Low', 'Medium', 'High'])

# Boxplot
boston_df.boxplot(column='MEDV', by='AGE_group')
plt.title('MEDV vs AGE Group')
plt.suptitle('')
plt.show()
```

```python
from scipy import stats
```

```python
chas0 = boston_df[boston_df['CHAS'] == 0]['MEDV']
chas1 = boston_df[boston_df['CHAS'] == 1]['MEDV']

levene_test = stats.levene(chas0, chas1)
levene_test
```

```python
ttest = stats.ttest_ind(chas0, chas1)
ttest
```

```python
low = boston_df[boston_df['AGE_group'] == 'Low']['MEDV']
med = boston_df[boston_df['AGE_group'] == 'Medium']['MEDV']
high = boston_df[boston_df['AGE_group'] == 'High']['MEDV']

anova = stats.f_oneway(low, med, high)
anova
```

```python
pearson_corr = stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])
pearson_corr
```

```python
import statsmodels.api as sm

X = boston_df['DIS']
y = boston_df['MEDV']

# thêm intercept
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
print(model.summary())
```
